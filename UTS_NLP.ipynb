1. Text Cleaning

```python
import re

# Input teks manual
text = "KPK Bakal Siapkan SP3 :) !!! Kunjungi https://kompas.com sekarang!!!"

# Fungsi text cleaning
def text_cleaning(text):
    # Hapus URL
    text = re.sub(r'http\S+|www\S+', '', text)

    # Hapus angka
    text = re.sub(r'\d+', '', text)

    # Hapus tanda baca
    text = re.sub(r'[^\w\s]', '', text)

    # Hapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Proses cleaning
clean_text = text_cleaning(text)

# Output hasil
print("Teks Asli  :", text)
print("Teks Bersih:", clean_text)
```

2. Tokenization

```python
# Input teks (biasanya hasil dari cleaning)
text = "KPK Bakal Siapkan SP Kunjungi sekarang"

# Fungsi tokenization
def tokenization(text):
    tokens = text.split()  # pisah berdasarkan spasi
    return tokens

# Proses tokenization
tokens = tokenization(text)

# Output hasil
print("Teks :", text)
print("Token:", tokens)
```

3. Normalization

```python
# Hasil dari tokenization
tokens = ['KPK', 'Bakal', 'Siapkan', 'SP', 'Kunjungi', 'sekarang']

# Kamus normalisasi (contoh sederhana)
normal_dict = {
    "kpk": "komisi pemberantasan korupsi",
    "sp": "surat perintah"
}

# Fungsi normalization
def normalization(tokens):
    normalized_tokens = []

    for token in tokens:
        token_lower = token.lower()  # samakan huruf kecil

        # cek di kamus
        if token_lower in normal_dict:
            normalized_tokens.append(normal_dict[token_lower])
        else:
            normalized_tokens.append(token_lower)

    return normalized_tokens

# Proses normalization
normalized = normalization(tokens)

# Output hasil
print("Token Asli      :", tokens)
print("Setelah Normalisasi:", normalized)
```

4. Stopword Removal

```python
# Hasil dari normalization
tokens = ['komisi pemberantasan korupsi', 'bakal', 'siapkan', 'surat perintah', 'kunjungi', 'sekarang']

# Daftar stopword sederhana (Bahasa Indonesia)
stopwords = [
    "dan", "di", "ke", "dari", "yang", "untuk", "pada",
    "dengan", "adalah", "ini", "itu", "sekarang"
]

# Fungsi stopword removal
def remove_stopwords(tokens):
    filtered_tokens = []

    for token in tokens:
        if token not in stopwords:
            filtered_tokens.append(token)

    return filtered_tokens

# Proses stopword removal
filtered = remove_stopwords(tokens)

# Output hasil
print("Sebelum:", tokens)
print("Sesudah:", filtered)
```

5. Case Folding

```python
# Input teks (sebelum case folding)
text = "KPK Bakal Siapkan SP Kunjungi Sekarang"

# Fungsi case folding
def case_folding(text):
    text = text.lower()  # ubah semua huruf jadi kecil
    return text

# Proses case folding
hasil = case_folding(text)

# Output
print("Sebelum:", text)
print("Sesudah:", hasil)
```

7. Lemmatization

```python
# Contoh token (hasil stemming sebelumnya)
tokens = ['komisi', 'berantas', 'korupsi', 'bakal', 'siap', 'surat', 'perintah']

# Kamus lemmatization sederhana
lemma_dict = {
    "berantas": "berantas",
    "siap": "siap",
    "korupsi": "korupsi"
}

# Fungsi lemmatization
def lemmatization(tokens):
    hasil = []

    for token in tokens:
        if token in lemma_dict:
            hasil.append(lemma_dict[token])
        else:
            hasil.append(token)

    return hasil

# Proses lemmatization
lemmatized = lemmatization(tokens)

# Output
print("Sebelum:", tokens)
print("Sesudah:", lemmatized)
```
